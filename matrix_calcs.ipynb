To show that there exists $C_{\textup{cond}} > 1$ so that for all $q \in (0,1)$, any $V \in \mathcal{V}$ and any $V' \in \mathcal{T}_{b} \cup \mathcal{L}_{b}$ it holds that
$$
    \mathsf{Pr}_{V}(W_{6} = V') \geq \frac{\pi_{\textup{bal}}(V')}{C_{\textup{cond}}}
$$
requires the calculation of the 6th power of $Q_{\text{bal}}$. Next, in order to prove that for $V' \in \mathcal{T}_{f} \cup \mathcal{L}_{f}$, 
$$
    \mathsf{Pr}_{V}(W_{i} = V' \text{ for some }i \in \{0,1,\dots,6\}) \geq \frac{\pi_{\textup{bal}}(V')}{C_{\textup{cond}}},
$$
we make use of the relation
$$
    \mathsf{Pr}_{V}(W_{i} = V' \text{ for some }i \in \{0,1,\dots,6\}) \geq \max\{\mathsf{Pr}_{V}(W_{5} = V'), \mathsf{Pr}_{V}(W_{6} = V')\}.
$$
To deal with the second bound, thus additionally requires analysis of the 5th power of $Q_{\text{bal}}$.

```python
import sympy as sp
```

This Python script uses **SymPy**, a symbolic mathematics library, to define and construct the 18x18 transition matrix $Q_{\textup{bal}}$ symbolically. The matrix represents a Markov chain, where each element describes transition probabilities between states. The states are
$$\{T_{b}^{(1)}, T_{b}^{(2)}, T_{b}^{(3)}, T_{f}^{(1)}, T_{f}^{(2)}, T_{f}^{(3)}, L_{b}^{(12)}, L_{b}^{(21)}, L_{b}^{(13)}, L_{b}^{(31)}, L_{b}^{(23)}, L_{b}^{(32)}, L_{f}^{(12)}, L_{f}^{(21)}, L_{f}^{(13)}, L_{f}^{(31)}, L_{f}^{(23)}, L_{f}^{(32)}\}$$
in this order. The script proceeds to compute the 5th and 6th powers of $Q_{\textup{bal}}$ in order to lower bound transition probabilities.

```python
q = sp.symbols('q')
Q_bal = [[0, q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0],
    [q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2)],
    [0, q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0],
    [q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2)],
    [q*sp.Rational(1,3), 0, 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0]]
Q_bal = sp.Matrix(Q_bal)
Q_bal
#print(sp.latex(Q))
```

Due to symmetry we can restrict our analysis to the columns representing $T_b^{(1)},T_f^{(1)},L_b^{(12)}$ and $L_f^{(12)}$ of the matrix $Q_\textup{bal}$, i.e., 1st, 4th, 7th and 13th columns. We first deal with the backward edge states $T_b^{(1)}$, and $L_b^{(12)}$ and analyze the 1st and 7th column of $Q_\textup{bal}^6$.

```python
Q_bal_6_sub_backward = Q_bal_6[:, [0, 6]].as_mutable()
Q_bal_6_sub_backward
#print(sp.latex(sub_Q_power_6))
```

Now we divide the first column of this sub matrix by $\pi_{\textup{bal}}(T_b^{(1)})$ and the second column by $\pi_{\textup{bal}}(L_b^{(12)})$, forgetting about constants, i.e., we divide every entry by $q$. What remains is a polynomial of degree $5$.

```python
sub_Q6_backward_polynomial = Q_bal_6_sub_backward.copy()
for i in range(18):
    sub_Q6_backward_polynomial[i, 0] = sp.together(sub_Q6_backward_polynomial[i, 0] / q)
    sub_Q6_backward_polynomial[i, 1] = sp.together(sub_Q6_backward_polynomial[i, 1] / q)
sub_Q6_backward_polynomial
```

Next, we need to check if each polynomial is strictly larger than $0$ in the interval $[0,1]$. We plot all polynomials to convince the reader of this claim. This immediately implies we can find a constant $C_1>0$ such that for each $V\in\mathcal{V}$ it holds that $\mathsf{Pr}_{V}(W_{6} = T_b^{(1)}) \geq \frac{\pi_{\textup{bal}}(T_b^{(1)})}{C_{1}}$ and $\mathsf{Pr}_{V}(W_{6} = L_b^{(12)}) \geq \frac{\pi_{\textup{bal}}(L_b^{(12)})}{C_{1}}$.

```python
p = sp.plotting.plot(sub_Q6_backward_polynomial[0,0], (q, 0, 1), show=False)

for i in range(18):
    for j in range(2):
        if i>0 or j>0:
            p.extend(sp.plotting.plot(sub_Q6_backward_polynomial[i,j], (q, 0, 1), show=False))
p.ylim = (0, 0.5)
p.axis_center = (0, 0)
p.ylabel = "Polynomial(q)"
p.show()
```

To conclude the proof a similar analysis has to be performed for the two forward edge states $T_f^{(1)}$ and $L_f^{(12)}$.
We need to consider both matrices $Q_\textup{bal}^5$ and $Q_\textup{bal}^6$ but can restrict our analysis to columns 4 and 13. We factorize the entries of the respective sub matrices of $Q_\textup{bal}^5$ and $Q_\textup{bal}^6$ by factoring out either $q$ or $1-q$. In the case of $Q_\textup{bal}^5$, we factor out $q$ when transitioning from a state in $\mathcal{T}$ or $\mathcal{L}$ to another state within the same subset after five time steps; otherwise, we factor out $1q$. For $Q_\textup{bal}^6$, the roles of $q$ and $1q$ are reversed.

```python
sub_Q_5_forward = Q_bal_5[:, [3, 12]].as_mutable()

for i in range(6, 18):
    sub_Q_5_forward[i, 0], remainder = sp.div(sub_Q_5_forward[i, 0], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_5_forward[i, 0] = sp.together(sub_Q_5_forward[i, 0])
    sub_Q_5_forward[i, 0] = (1 - q) * sub_Q_5_forward[i, 0]
for i in range(6):
    sub_Q_5_forward[i, 1], remainder = sp.div(sub_Q_5_forward[i, 1], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_5_forward[i, 1] = sp.together(sub_Q_5_forward[i, 1])
    sub_Q_5_forward[i, 1] = (1 - q) * sub_Q_5_forward[i, 1]

sub_Q_5_forward
#print(sp.latex(sub_Q_power_5))
```

```python
Q_bal_6_sub_forward = Q_bal_6[:, [3, 12]].as_mutable()

for i in range(6):
    Q_bal_6_sub_forward[i, 0], remainder = sp.div(Q_bal_6_sub_forward[i, 0], 1 - q)
    if remainder != 0:
        print('Error')
    Q_bal_6_sub_forward[i, 0] = sp.together(Q_bal_6_sub_forward[i, 0])
    Q_bal_6_sub_forward[i, 0] = (1 - q) * Q_bal_6_sub_forward[i, 0]
for i in range(6,18):
    Q_bal_6_sub_forward[i, 1], remainder = sp.div(Q_bal_6_sub_forward[i, 1], 1 - q)
    if remainder != 0:
        print('Error')
    Q_bal_6_sub_forward[i, 1] = sp.together(Q_bal_6_sub_forward[i, 1])
    Q_bal_6_sub_forward[i, 1] = (1 - q) * Q_bal_6_sub_forward[i, 1]

Q_bal_6_sub_forward
#print(sp.latex(sub_Q_power_6))
```

Since for $V' \in\mathcal{T}_f \cup \mathcal{L}_f$ it holds that $\pi_\text{Bal}(V') = O(1-q)$, we lower bound each entry of entrywise maximum (to be understood entrywise) of the sub matrices of $Q_\textup{Bal}^5$ and $Q_\textup{Bal}^6$ with the respective entry that contains the factor $1-q$.

```python
lower_bound_max_Q_bal_5_Q_bal_6_forward = sp.Matrix.hstack(sp.Matrix.vstack(Q_bal_6_sub_forward[0:6, 0], sub_Q_5_forward[6:18, 0]), sp.Matrix.vstack(sub_Q_5_forward[0:6, 1], Q_bal_6_sub_forward[6:18, 1]))
lower_bound_max_Q_bal_5_Q_bal_6_forward
#print(sp.latex(lower_bound_matrix))
```

Now we can repeat the analysis done for the forward edge state case.

```python
lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial = lower_bound_max_Q_bal_5_Q_bal_6_forward.copy()
for i in range(18):
    lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[i, 0] = sp.together(lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[i, 0] / (1 - q))
    lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[i, 1] = sp.together(lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[i, 1] / (1 - q))

lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial
```

```python
p = sp.plotting.plot(lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[0,0], (q, 0, 1), show=False)

for i in range(18):
    for j in range(2):
        if i>0 or j>0:
            p.extend(sp.plotting.plot(lower_bound_max_Q_bal_5_Q_bal_6_forward_polynomial[i,j], (q, 0, 1), show=False))
p.ylim = (0, 0.5)
p.axis_center = (0, 0)
p.show()
```
