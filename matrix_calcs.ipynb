### Proof of Lemma 4.1

```python
import sympy as sp
```

This Python script uses **SymPy**, a symbolic mathematics library, to define and construct a 18x18 **transition matrix** $Q$ symbolically. The matrix represents a **Markov chain**, where each element describes transition probabilities between states. The script proceeds to compute the 5th and 6th powers of $Q$ in order to lower bound transition probabilities.

```python
q = sp.symbols('q')
Q = [[0, q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0],
    [q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2)],
    [0, q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2), 0, 0],
    [q*sp.Rational(1,3), q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0, q*sp.Rational(1,6), q*sp.Rational(1,6), 0, 0, 0, 0, sp.Rational(1,2) - q*sp.Rational(1,2), sp.Rational(1,2) - q*sp.Rational(1,2)],
    [q*sp.Rational(1,3), 0, 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 0, 0, 1 - q, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0, 0],
    [0, q*sp.Rational(1,3), 0, 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 0, 1 - q, 0, q*sp.Rational(1,3), 0, 0, 0, 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0],
    [0, 0, q*sp.Rational(1,3), 1 - q, 0, 0, 0, 0, q*sp.Rational(1,3), 0, q*sp.Rational(1,3), 0, 0, 0, 0, 0, 0, 0]]
Q = sp.Matrix(Q)
Q
#print(sp.latex(Q))
```

```python
Q_power_5 = sp.simplify(Q**5)
Q_power_5
```

```python
Q_power_6 = sp.simplify(Q**6)
Q_power_6
```

Due to symmetry we can restrict our analysis to the 1st, 4th, 7th and 13th columns of the matrix Q. Since we only need to consider $Q^5$ when transitioning into a state with a forward edge, we can even restrict ourselves to column 4 and 13 for the analysis $Q^5$. We factorize the entries of the respective sub matrices of $Q^5$ and $Q^6$ by factoring out either $q$ or $1-q$.

```python
sub_Q_power_5 = Q_power_5[:, [3, 12]].as_mutable()

for i in range(6, 18):
    sub_Q_power_5[i, 0], remainder = sp.div(sub_Q_power_5[i, 0], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_power_5[i, 0] = sp.together(sub_Q_power_5[i, 0])
    sub_Q_power_5[i, 0] = (1 - q) * sub_Q_power_5[i, 0]
for i in range(6):
    sub_Q_power_5[i, 1], remainder = sp.div(sub_Q_power_5[i, 1], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_power_5[i, 1] = sp.together(sub_Q_power_5[i, 1])
    sub_Q_power_5[i, 1] = (1 - q) * sub_Q_power_5[i, 1]

sub_Q_power_5
#print(sp.latex(sub_Q_power_5))
```

```python
sub_Q_power_6 = Q_power_6[:, [0, 3, 6, 12]].as_mutable()

for i in range(6):
    sub_Q_power_6[i, 1], remainder = sp.div(sub_Q_power_6[i, 1], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_power_6[i, 1] = sp.together(sub_Q_power_6[i, 1])
    sub_Q_power_6[i, 1] = (1 - q) * sub_Q_power_6[i, 1]
for i in range(6,18):
    sub_Q_power_6[i, 3], remainder = sp.div(sub_Q_power_6[i, 3], 1 - q)
    if remainder != 0:
        print('Error')
    sub_Q_power_6[i, 3] = sp.together(sub_Q_power_6[i, 3])
    sub_Q_power_6[i, 3] = (1 - q) * sub_Q_power_6[i, 3]

sub_Q_power_6
#print(sp.latex(sub_Q_power_6))
```

Choosing either the entry of $Q^5$ or $Q^6$ depending on whether the factor aligns with $\pi_\text{Bal}(V)$, where $V$ is the state that aligns with the row number, we construct a matrix that lower bounds the transition probabilities.

```python
lower_bound_matrix = sp.Matrix.hstack(sub_Q_power_6[:,0], sp.Matrix.vstack(sub_Q_power_6[0:6, 1], sub_Q_power_5[6:18, 0]), sub_Q_power_6[:, 2], sp.Matrix.vstack(sub_Q_power_5[0:6, 1], sub_Q_power_6[6:18, 3]))
lower_bound_matrix
#print(sp.latex(lower_bound_matrix))
```

In the next step we delete the factor of each entry.

```python
lower_bound_matrix_polynomial = lower_bound_matrix.copy()
for i in range(18):
    lower_bound_matrix_polynomial[i, 0] = sp.together(lower_bound_matrix_polynomial[i, 0] / q)
    lower_bound_matrix_polynomial[i, 2] = sp.together(lower_bound_matrix_polynomial[i, 2] / q)
    lower_bound_matrix_polynomial[i, 1] = sp.together(lower_bound_matrix_polynomial[i, 1] / (1 - q))
    lower_bound_matrix_polynomial[i, 3] = sp.together(lower_bound_matrix_polynomial[i, 3] / (1 - q))

lower_bound_matrix_polynomial
```

Finally, we calculate the value of each entry at the boundary, i.e., if $q$ is 0 or 1. We confirm that all entries are positive. Since all entries are continuous functions (polynomials) and for each $q\in(0,1)$ there is a positive probability to end up any state starting from any state after 5 or 6 runs of the Markov chain (proof left for the reader; the plots below indicate the behavior), the statement of the lemma follows. 

```python
lower_bound_matrix_polynomial_zero = lower_bound_matrix_polynomial.copy()
lower_bound_matrix_polynomial_one = lower_bound_matrix_polynomial.copy()
for i in range(18):
        for j in range(4):
                lower_bound_matrix_polynomial_zero[i,j] = lower_bound_matrix_polynomial[i,j].subs(q,0)
                lower_bound_matrix_polynomial_one[i,j] = lower_bound_matrix_polynomial[i,j].subs(q,1)
                if lower_bound_matrix_polynomial_zero[i,j] < 0 or lower_bound_matrix_polynomial_one[i,j] < 0:
                        print("Error")

print(lower_bound_matrix_polynomial_zero)
print(lower_bound_matrix_polynomial_one)
#print(sp.latex(lower_bound_matrix_polynomial_zero))
#print(sp.latex(lower_bound_matrix_polynomial_one))
```
